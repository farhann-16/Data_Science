```python
import numpy as np
import time

size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x + y for x,y in zip(l1,l2)]
end = time.time()

print(end - start)
```

```python
#Using the numpy arrays 

size = 1_000_000

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()                    #Record the st time
add = l1 + l2
end = time.time()                      #Record the end time

print(end - start)                     #Return the end time - st time
```
