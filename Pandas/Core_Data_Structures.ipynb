# Core Data Structures in Pandas

Pandas is built on two main data structures:

1. **Series** â†’ One-dimensional (like a single column in Excel)  
2. **DataFrame** â†’ Two-dimensional (like a full spreadsheet or SQL table)

---

## ðŸ“Š Series â€” 1D Labeled Array

A **Series** is like a list with labels (index).

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40])
print(s)
```
**Output:**
```
0    10
1    20
2    30
3    40
dtype: int64
```

Notice the **automatic index**: `0, 1, 2, 3`

You can also define a **custom index**:

```python
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(s)
```
**Output:**
```
a    10
b    20
c    30
dtype: int64
```

---

### ðŸ” Series vs Dictionary

A `pandas.Series` may look similar to a Python **dictionary** because both store data with labels, but a **Series** offers much more:

| Feature | Python Dictionary | Pandas Series |
|----------|-------------------|----------------|
| Data structure | Keyâ€“Value store | Labeled 1D array |
| Operations | Manual looping | Vectorized (fast) |
| Missing data | Must handle manually | Supports `NaN` |
| Indexing | Keys only | Labels + Positions |
| Integration | Limited | Works with DataFrames, NumPy, etc. |

âœ… **In short:** A Series is better for **data analysis** because it supports fast arithmetic, automatic index alignment, and flexible data access.

## ðŸ§® DataFrame â€” 2D Labeled Table

A **DataFrame** is like a **dictionary of Series** â€” multiple columns with labels.

```python
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "city": ["Delhi", "Mumbai", "Bangalore"]
}

df = pd.DataFrame(data)
print(df)
```
**Output:**
```
      name  age       city
0    Alice   25      Delhi
1      Bob   30     Mumbai
2  Charlie   35  Bangalore
```

Each **column** in a DataFrame is a **Series**.

You can access columns like this:

```python
print(df["name"])  # Access a Series
```

---

## ðŸ·ï¸ Index and Labels

Every **Series** and **DataFrame** has an **Index**, which helps with:

- Fast lookups  
- Aligning data  
- Merging & joining  
- Time series operations

You can check them using:

```python
df.index      # Row labels
df.columns    # Column labels
```

You can also change them:

```python
df.index = ["a", "b", "c"]
df.columns = ["Name", "Age", "City"]
print(df)
```
**Output:**
```
   Name  Age       City
a  Alice  25      Delhi
b    Bob  30     Mumbai
c Charlie 35  Bangalore
```

---

## ðŸŽ¯ Why Learn These Well?

Most Pandas operations are built on these two foundations:

- Selection (`df["col"]` or `df.loc[]`)  
- Filtering (`df[df["age"] > 30]`)  
- Merging (`pd.merge(df1, df2)`)  
- Aggregation (`df.groupby("col").mean()`)  

Understanding **Series** and **DataFrames** will make all future Pandas tasks much easier.
