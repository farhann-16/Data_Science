# Aggregation & Grouping in Pandas

Grouping and aggregating helps you **summarize your data** â€” like answering:
- â€œWhatâ€™s the **average salary per department**?â€  
- â€œHow many **users joined per month**?â€

Pandas makes this easy with the powerful `.groupby()` method.

---

## ðŸ‘¥ The `.groupby()` Function

`df.groupby()` is used to group rows of a DataFrame based on values in one or more columns,  
allowing you to perform **aggregate functions** like `sum()`, `mean()`, `count()`, etc.

### Example DataFrame

```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})

print(df)
```

---

### ðŸ§® Group by a Single Column

```python
df.groupby("Department")["Salary"].mean()
```
> â€œGroup by Department, then calculate the **average Salary** for each group.â€

---

### âš™ï¸ Common Aggregation Functions

```python
df.groupby("Team")["Salary"].mean()   # Average per team
df.groupby("Team")["Salary"].sum()    # Total salary per team
df.groupby("Team")["Salary"].count()  # Number of records per team
df.groupby("Team")["Salary"].min()    # Minimum salary per team
df.groupby("Team")["Salary"].max()    # Maximum salary per team
```

---

### ðŸ“Š Group by Multiple Columns

```python
df.groupby(["Team", "Gender"])["Salary"].mean()
```
> Calculates average salary for each **Teamâ€“Gender combination**.

---

## ðŸ§  Custom Aggregations with `.agg()`

You can apply **multiple functions at once** to grouped data.

```python
df.groupby("Team")["Salary"].agg(["mean", "max", "min"])
```

ðŸŸ¢ `.agg()` and `.aggregate()` are **the same function** â€” theyâ€™re just aliases.

---

### ðŸ·ï¸ Name Your Own Aggregation Columns

```python
df.groupby("Team")["Salary"].agg(
    avg_salary="mean",
    high_salary="max"
)
```

---

### ðŸ§© Apply Different Functions to Different Columns

```python
df.groupby("Team").agg({
    "Salary": "mean",
    "Age": "max"
})
```

This gives the **average Salary** and **maximum Age** for each team.

---

## ðŸ” Transform vs Aggregate vs Filter

| Operation | Returns | When to Use | Example |
|------------|----------|--------------|----------|
| `.aggregate()` | Single value per group | Summary statistic (e.g., mean) | `df.groupby("Team")["Salary"].mean()` |
| `.transform()` | Same shape as original | Add new derived columns | `df["Team Avg"] = df.groupby("Team")["Salary"].transform("mean")` |
| `.filter()` | Subset of rows | Keep or discard groups | `df.groupby("Team").filter(lambda x: x["Salary"].mean() > 80)` |

---

### ðŸ§® `.transform()` Example

Adds a **new column** showing the average salary of each employeeâ€™s team:

```python
df["Team Avg"] = df.groupby("Team")["Salary"].transform("mean")
print(df[["Team", "Salary", "Team Avg"]])
```

**Output:**
```
  Team  Salary  Team Avg
0    A      85      87.5
1    A      90      87.5
2    B      78      81.5
3    B      85      81.5
4    C      92      90.0
5    C      88      90.0
6    D      75      77.5
7    D      80      77.5
```

---

### ðŸ” `.filter()` Example

Keep only the teams whose **average salary > 80**:

```python
df_filtered = df.groupby("Team").filter(lambda x: x["Salary"].mean() > 80)
print(df_filtered)
```

---

## âœ… Summary

| Concept | Function | Description |
|----------|-----------|-------------|
| **Group by one or more columns** | `df.groupby()` | Splits data into groups |
| **Compute statistics** | `.mean()`, `.sum()`, `.count()` | Common aggregate functions |
| **Multiple aggregations** | `.agg()` | Apply multiple or custom metrics |
| **Per-row group values** | `.transform()` | Adds computed group stats to original data |
| **Filter groups** | `.filter()` | Keeps groups that meet conditions |

---

### ðŸ’¡ Key Takeaways

- `.groupby()` is your **go-to tool** for summarizing large datasets.  
- Use `.agg()` for **custom or multiple aggregations**.  
- Use `.transform()` to **add new columns** with group-level info.  
- Use `.filter()` to **select only certain groups** based on conditions.
