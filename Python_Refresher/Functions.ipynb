## ğŸ”¹ 2. Check Prime Number

```python
def is_prime(n):
    """Check if a number is prime."""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example
print(is_prime(7))  # Output: True
```

**Explanation:**  
This function checks if a number is prime by dividing it with all integers up to its square root. Returns `True` if prime, otherwise `False`.
