# Data Selection & Filtering in Pandas

Selecting the right **rows and columns** is the first step in analyzing any dataset.  
Pandas provides several **powerful and flexible** ways to access and filter data efficiently.

---

## ðŸŽ¯ Selecting Rows & Columns

### Selecting Columns

```python
df["column_name"]      # Single column (as Series)
df[["col1", "col2"]]   # Multiple columns (as DataFrame)
```

Example:
```python
df["Name"]
df[["Name", "Age"]]
```

---

## ðŸ”¢ Selecting Rows by Index

Pandas provides two primary methods:

- **`.loc[]` â†’ Label-based access** (uses row/column labels)
- **`.iloc[]` â†’ Position-based access** (uses integer positions)

```python
df.loc[0]     # First row (by label)
df.iloc[0]    # First row (by position)
```

---

## ðŸŽ¯ Selecting Specific Rows and Columns

```python
df.loc[0, "Name"]   # Value at row 0, column 'Name'
df.iloc[0, 1]       # Value at row 0, column at index 1
```

### Slicing Rows and Columns

```python
df.loc[0:2, ["Name", "Age"]]  # Rows 0 to 2, specific columns
df.iloc[0:2, 0:2]             # Rows and columns by position
```

---

## âš¡ Fast Access: `.at[]` and `.iat[]`

These methods are optimized for **single-element** access (faster than `.loc` or `.iloc`).

```python
df.at[0, "Name"]   # Fast label-based access
df.iat[0, 1]       # Fast position-based access
```

---

## ðŸ” Filtering Data with Conditions

### Simple Condition

```python
df[df["Age"] > 30]
```

### Multiple Conditions (AND / OR)

```python
df[(df["Age"] > 25) & (df["City"] == "Delhi")]
df[(df["Name"] == "Bob") | (df["Age"] < 30)]
```

ðŸ§  **Tip:** Always use **parentheses** around each condition when combining them.

---

## ðŸ§¾ Querying Data with `.query()`

The `.query()` method provides a **clean, SQL-like syntax** for filtering rows.  
Itâ€™s more **readable and concise** than traditional boolean indexing.

### Example:

```python
df.query("Age > 25 and City == 'Delhi'")
```

You can even use dynamic variables:

```python
col = "Age"
df.query(f"{col} > 25")
```

---

## ðŸ§  Rules and Tips for Using `.query()`

### 1. Column names become variables
You can reference them directly:

```python
df.query("age > 25 and city == 'Delhi'")
```

### 2. String values must be in quotes
```python
df.query("name == 'Harry'")
df.query('city == "Mumbai"')
```

### 3. Use backticks for columns with spaces or special characters
```python
df.query("`first name` == 'Alice'")
```

### 4. Use `@` to reference Python variables
```python
age_limit = 30
df.query("age > @age_limit")
```

### 5. Logical Operators
| Operator | Use in `.query()` | Example |
|-----------|------------------|----------|
| `and` | Logical AND | `age > 30 and city == 'Delhi'` |
| `or` | Logical OR | `city == 'Delhi' or city == 'Mumbai'` |
| `not` | Negation | `not (age > 40)` |

âŒ **Bad:**  
```python
df.query("age > 30 & city == 'Delhi'")
```
âœ… **Good:**  
```python
df.query("age > 30 and city == 'Delhi'")
```

### 6. Chained Comparisons
```python
df.query("25 < age <= 40")
```

### 7. Avoid reserved keywords as column names
If you have columns like `class`, `lambda`, etc., use backticks:
```python
df.query("`class` == 'Physics'")
```

### 8. Case Sensitivity
Both column names and string values are **case-sensitive**:
```python
df.query("City == 'delhi'")  # âŒ if actual value is 'Delhi'
```

### 9. `.query()` returns a copy, not a view
```python
filtered = df.query("age < 50")
```

Changes to `filtered` **wonâ€™t affect** the original `df` unless reassigned.
