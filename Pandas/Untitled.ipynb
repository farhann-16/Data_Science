```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("p2.csv")
df
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
df['Purchase_Amount'] = df['Purchase_Amount'].fillna(df['Purchase_Amount'].median())
df
```

- Remove extra spaces from Name, City, and Email.

```python
col = ['Name', 'City', 'Email']                                 # Variable = Not''
df[col] = df[col].apply(lambda col: col.str.strip())            # Use strip when removing space 
df
```

- Convert all City values to title case (e.g., Delhi, Mumbai).

```python
df['City'] = df['City'].str.title()
df
```

##### Q3. Standardize Gender

- Fix inconsistent gender labels:
- "M", "F", "f" â†’ should become Male, Female

```python
df['Gender'] = df['Gender'].str.strip().str.upper().replace({                     # We can replace values more than 1 by direct dic type value
    "M" : 'Male',                                                                 # Strip for removing space 
    "F" : 'Female'})
df
```

```python
df['Age'] = df['Age'].astype(int)
df                                      # Convert age into int
```
