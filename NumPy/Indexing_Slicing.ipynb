## ðŸ”¹ Indexing (Same as Python Lists)

```python
import numpy as np

arr = np.array([10, 20, 30, 40])

print(arr[0])   # 10 (first element)
print(arr[-1])  # 40 (last element)
```

You can use **positive** or **negative** indices to access elements in a NumPy array.

---

## ðŸ”¹ Slicing (Extracting Parts of an Array)

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[1:4])   # [20 30 40] â†’ slice from index 1 to 3
print(arr[:3])    # [10 20 30] â†’ first 3 elements
print(arr[::2])   # [10 30 50] â†’ every 2nd element
```

ðŸ§  **Note:**  
Slicing in NumPy returns a **view**, not a copy.  
This means both arrays share the **same memory** â€” changes in one affect the other.

---

## âš™ï¸ Example: Slicing Returns a View

```python
arr = np.array([10, 20, 30, 40, 50])

sliced = arr[1:4]
sliced[0] = 999

print(arr)
# Output: [10 999 30 40 50]
```

As you can see, changing `sliced` also changed the original array!

---

## ðŸ§  Why Does This Happen?

- **Memory Efficiency:** Avoids unnecessary copies, saving memory.  
- **Performance:** Faster access and manipulation of large datasets.

---

## ðŸ› ï¸ Creating a Copy (Independent Array)

If you want a **separate copy** of the data that wonâ€™t affect the original array:

```python
arr = np.array([10, 20, 30, 40, 50])
copy_arr = arr[1:4].copy()

copy_arr[0] = 999
print(arr)       # [10 20 30 40 50] (unchanged)
print(copy_arr)  # [999 30 40]
```

âœ… Use `.copy()` whenever you want an **independent** array.

---

**Summary:**
- Indexing gives individual elements.
- Slicing extracts parts of an array.
- NumPy slicing returns **views** (shared memory).
- Use `.copy()` to make **independent copies**.
